```python
img = cv.imread('/content/drive/MyDrive/20211222_102527.jpg')
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
rotated_img = cv.rotate(img2, cv.ROTATE_90_COUNTERCLOCKWISE)


start_point = (900, 1200)
end_point = (2400, 1500)


color = (255, 255, 255)
thickness = -1

image_with_rectangle = cv.rectangle(rotated_img.copy(), start_point, end_point, color, thickness)

center_coordinates = (1200, 800)
radius = 250

color = (255, 0, 255)
thickness = -1

image_with_rectangle2 = cv.circle(image_with_rectangle.copy(), center_coordinates, radius, color, thickness)

plt.imshow(image_with_rectangle2)
```
